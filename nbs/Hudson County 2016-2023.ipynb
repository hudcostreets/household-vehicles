```python
from utz import *
from hhvs import *
ire.chdir('../hudson')
```

```python
d = YearsDF()
d
```

```python
d['Total vehicles'] = d['1 vehicle'] + d['2 vehicles']*2 + d['3 vehicles']*3 + d['4+ vehicles']*4
d['1st vehicles'] = d['4+ vehicles'] + d['3 vehicles'] + d['2 vehicles'] + d['1 vehicle']
d['2nd vehicles'] = d['4+ vehicles'] + d['3 vehicles'] + d['2 vehicles']
d['3rd vehicles'] = d['4+ vehicles'] + d['3 vehicles']
d['4th+ vehicles'] = d['4+ vehicles']
d
```

```python
d['Vehicles per household'] = d['Total vehicles'] / d['Total households']
d
```

```python
d.reset_index(level=0).loc['Jersey City']
```

```python
c = d.reset_index(level=0).reset_index(drop=True).drop(columns='Vehicles per household').groupby('year').sum()
c['Vehicles per household'] = c['Total vehicles'] / c['Total households']
c
```

```python
pcts = c[['0 vehicles', '1 vehicle', '2 vehicles', '3 vehicles', '4+ vehicles']].div(c['Total households'], 0)
pcts.columns.name = 'Vehicles per household'
pcts
```

```python
pct_plot(
    pcts,
    title='Hudson County – households by vehicle count',
    melt='% households',
    name='hc_vehs_years',
)
```

```python
vs = c[['1st vehicles', '2nd vehicles', '3rd vehicles', '4th+ vehicles']]
vs.columns.name = 'Vehicle #'
vs
```

```python
botleg = dict(
    x=.5,
    xanchor='center',
    yanchor='top',
    y=-.05,
    orientation='h',
    title_text='',
)
```

```python
abs_plot(
    vs,
    title='Hudson County – total vehicles',
    subtitle='by household vehicle #',
    melt='Vehicles',
    name='hc_vehs',
    colors=colors[1:],
    xaxis=dict(dtick=1, title_text=''),
    x_tickangle=0,
    legend=botleg,
    layout=dict(margin=dict(t=0, l=0, r=10, b=0)),
)
```

```python
vp = c[['1st vehicles', '2nd vehicles', '3rd vehicles', '4th+ vehicles']]
vp = vp.div(vp.sum(1), 0)
vp.columns.name = '% Vehicles'
vp
```

```python
pct_plot(
    vp,
    title="Hudson County – vehicle %'s",
    subtitle='by household vehicle #',
    melt='% households',
    x_tickangle=0,
    legend=botleg,
    name='hc_vehs_pcts',
)
```
